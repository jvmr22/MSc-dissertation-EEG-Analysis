# Simple sine wave

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
time = np.arange(0, 2*math.pi, math.pi/100)
time
```

```python
amplitude = np.sin(time)
amplitude
```

```python
plt.plot(time, amplitude)
plt.show()
```

# Multiple sine waves
https://pythonnumericalmethods.berkeley.edu/notebooks/chapter24.01-The-Basics-of-waves.html

## example 1

```python
# sampling rate
sr = 500.0
# sampling interval
ts = 1.0/sr
t = np.arange(0,1,ts)
```

```python
freq_1 = 15   
y_1 = 5*np.sin(2*np.pi*freq_1*t)

plt.figure(figsize = (8, 8))
plt.subplot(311)
plt.plot(t, y_1, 'b')
plt.ylabel('Amplitude')

freq_2 = 10
y_2 = 10*np.sin(2*np.pi*freq_2*t + 10)

plt.subplot(312)
plt.plot(t, y_2, 'b')
plt.ylabel('Amplitude')

plt.xlabel('Time (s)')

y_3 = y_1+y_2

plt.subplot(313)
plt.plot(t, y_3, 'b')
plt.ylabel('Amplitude')

plt.xlabel('Time (s)')

plt.show()

```

## Example 2

```python
# sampling rate
sr = 100
# sampling interval
ts = 1.0/sr
t = np.arange(0,1,ts)
```

```python

freq = 1.
x = 3*np.sin(2*np.pi*freq*t)

plt.figure(figsize = (8, 8))
plt.subplot(411)
plt.plot(t, x, 'r')
plt.ylabel('Amplitude')


freq = 4
x += np.sin(2*np.pi*freq*t)

plt.subplot(412)
plt.plot(t, x, 'b')
plt.ylabel('Amplitude')

freq = 7   
x += 0.5* np.sin(2*np.pi*freq*t)

plt.subplot(413)
plt.plot(t, x, 'g')
plt.ylabel('Amplitude')

#plt.plot(t, x, 'r')
#plt.ylabel('Amplitude')

plt.show()
```
